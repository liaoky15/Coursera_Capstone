```python
radius = 1500
search_query = 'gym'
all_gyms = pd.DataFrame()
dist = []

for x in range(12):
    dist_lat = df.loc[x, 'latitude'] # neighborhood latitude value
    dist_lng = df.loc[x, 'longitude'] # neighborhood longitude value

    dist_name = df.loc[x, 'District Name']
    
    
    url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(
            CLIENT_ID,
            CLIENT_SECRET,
            dist_lat,
            dist_lng,
            VERSION,
            search_query,
            radius,
            LIMIT)
    
    results = requests.get(url).json()
    venues = results['response']['venues']
    
    nearby_gyms = json_normalize(venues)
    filtered_columns = ['id', 'name', 'location.lat', 'location.lng']
    nearby_gyms = nearby_gyms.loc[:, filtered_columns]
    nearby_gyms.columns = ['Gym ID','Gym Name', 
                  'Gym Latitude', 
                  'Gym Longitude']
    
    all_gyms = all_gyms.append( nearby_gyms, ignore_index = True)
    dist.append(dist_name)
    

print(dist)
all_gyms['District Name'] = dist    
```

```python
final_map = folium.Map(location=[25.0630, 121.5654], zoom_start=12.3) 

lati = all_gyms['Gym Latitude']
lngo = all_gyms['Gym Longitude']
labels = all_gyms['Gym Name']

for i in range(71):
    folium.Marker([lati[i],lngo[i]], popup = folium.Popup(labels[i],parse_html=True)).add_to(final_map)
    
final_map
```
